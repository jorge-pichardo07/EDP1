b) Calculando para $U(x)= e^{-x}$ para los coeficientes de fourier ocupamos la formula que vimos en clase

```python
from sympy import *
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from pylab import *

def ex(x):
    y = zeros(len(x))
    for i in range(len(x)):
        y[i] =  np.exp(-x[i])
    return y

def fourier(x1,N=10):
    J = zeros(len(x1))
    for n in range(1,N):
        J += 2*(1-(-1)**n)*np.sin(n*np.pi*x1)/n
    return (1.0/pi)*J

x= linspace(-1,1,100)
f = ex(x)
x1= linspace(-1,1,100)
E= fourier(x1)
plot(x,f,label='$f(x)= 0$')
plot(x1,f*E,label='Serie de Fourier')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Funci√≥n generalizada")
xlim([0.00,1.00]),ylim([0.00,1.00]),grid(),legend()

show()
```
