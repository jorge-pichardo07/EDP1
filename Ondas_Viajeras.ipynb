Graficar la función $f(x´)= e^{- \frac{(x´-0.5)^2}{0.001}}$



```python

import numpy as np
import matplotlib.pyplot as plt

def f(x_1):
    return np.exp(-(x_1 - 0.5)**2 / 0.01)


x_1 = np.linspace(0, 1, 1000)

y = f(x_1)

plt.figure(figsize=(8, 5))
plt.plot(x_1, y, 'black', linewidth=2)
plt.title('Perturbación inicial en la cuerda\n$f(x\') = \exp\\left(-\\frac{(x\' - 0.5)^2}{0.01}\\right)$')
plt.xlabel('Posición (x\')')
plt.ylabel('Amplitud de la perturbación')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xlim(0, 1)
plt.ylim(0, 1.1)

plt.axvline(x=0.5, color='r', linestyle='--', alpha=0.5)
plt.text(0.5, -0.1, 'x\' = 0.5', ha='center', va='top', color='r')

plt.show()
```

Aquí nombré a la función ($x´$)  como  $x_1$




```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(0, 1, 500)
c = 1
t = 0.2  # instante de tiempo
sigma = 0.01
x0 = 0.5


def f(xp):
    return np.exp(-((xp - x0)**2) / sigma)


fa = f(x - c*t)
fb = f(x + c*t)
fc = fa + fb


plot_config = {
    'figsize': (8, 5),
    'grid': True,
    'xlabel': 'Posición (x)',
    'ylabel': 'Amplitud',
    'xlim': (0, 1),
    'ylim': (0, 1.1)
}


plt.figure(figsize=plot_config['figsize'])
plt.plot(x, fa, 'b-', linewidth=2, label=f'f(x - ct), t = {t}')
plt.title('Perturbación moviéndose a la derecha\n$f(x\') = \exp\\left(-\\frac{(x\' - 0.5)^2}{0.01}\\right)$\n$x\' = x - ct$')
plt.xlabel(plot_config['xlabel'])
plt.ylabel(plot_config['ylabel'])
plt.xlim(plot_config['xlim'])
plt.ylim(plot_config['ylim'])
plt.grid(plot_config['grid'])
plt.legend()
plt.show()

plt.figure(figsize=plot_config['figsize'])
plt.plot(x, fb, 'r-', linewidth=2, label=f'f(x + ct), t = {t}')
plt.title('Perturbación moviéndose a la izquierda\n$f(x\') = \exp\\left(-\\frac{(x\' - 0.5)^2}{0.01}\\right)$\n$x\' = x + ct$')
plt.xlabel(plot_config['xlabel'])
plt.ylabel(plot_config['ylabel'])
plt.xlim(plot_config['xlim'])
plt.ylim(plot_config['ylim'])
plt.grid(plot_config['grid'])
plt.legend()
plt.show()

plt.figure(figsize=plot_config['figsize'])
plt.plot(x, fc, 'k-', linewidth=2, label='Superposición f(x - ct) + f(x + ct)')
plt.title('Superposición de ondas viajeras\nt = {0}'.format(t))
plt.xlabel(plot_config['xlabel'])
plt.ylabel(plot_config['ylabel'])
plt.xlim(plot_config['xlim'])
plt.ylim(0, 2.2)
plt.grid(plot_config['grid'])
plt.legend()
plt.show()
```

Aqui se muestra como la onda se va hacia la derecha con el valor de t=0.2, pasa lo mismo en la posicion de la derecha, en la superposicion se nota ambos lados, el signo hace que se mueva de lado
