En la siguiente celda, hice una función para solo insertar los datos, y el número de datos, junto con un nivel de confianza, donde D: la lista generada por nosotros de los datos (o la que se nos dé), n: número de la muestra q nos dan, c: el nivel de confianza, al final hago una comparación del estadistico z, con el estadistico de la normal en el nivel de significancia

```python
#Funcion que ocupa, una lista de datos, numero de datos y nuvel de confianza, para funcion uniforme
def Promedios(D, n, C):
  M = np.mean(D)
  z_0 = ( M-0.5)*(np.sqrt(n))/np.sqrt(1/12)
  z_n = stats.norm.ppf(C)
  if z_0 < z_n:
    a = "Aceptamos la hipòtesis, con un 95% de confianza, la media de los datos es igual a 1/2"
  else:
      a = "Rechazamos h0, no hay evidencia suficiente de que la media sea igual a 1/2";
  return a
```
